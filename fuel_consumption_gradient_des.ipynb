## Creating the model


```python
x_ens = mydata.ENGINESIZE
ys=mydata.CO2EMISSIONS
#here, both are in series so we have to convert them into array because regression needs array

#Converting series into numpy array
x_en = np.array(x_ens)
y = np.array(ys)

from sklearn.model_selection import train_test_split
```

#### splitting in train and test modules

```python
x_train,x_test,y_train,y_test=train_test_split(x_en,y,test_size=.30)
# 0.30 means 30% split of data
```

```python
print('samples are : x TRAIN {} and x test {}'.format(len(x_train),len(x_test)))
print('            : y TRAIN {} and y test {}'.format(len(y_train),len(y_test)))
```

```python
#reshaping to convert train and test in 2D arrays beacuse regression only accepts 2D arrays
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
teacher = LinearRegression()
learner = teacher.fit(x_train,y_train)
```

```python
yp = learner.predict(x_test)
```

```python
#printing intercepts and coefficient
c = learner.intercept_
m = learner.coef_
print("c is {} and\nm is {} ".format(c,m))

```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(yp,y_test)
error = np.sqrt(error)
```

```python
import seaborn as sns
#number of plots
s,axis = plt.subplots(1,2)
#set figure size
sns.set(rc={'figure.figsize':(7,5)})
#plotting each feature
sns.regplot(x=x_test,y=y_test,data=mydata,ax=axis[0]).set_title('actual v/s value')
sns.regplot(x=x_test,y=yp,data=mydata,ax=axis[1]).set_title('predicted v/s value ')
print('MSE '+ str(error)+'\n')
plt.show()
```
