```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv('DataSets-master/data_brain_own_model_design.csv')
```

```python
mydata.info
```

```python
x_input=mydata.iloc[:,0:4]
y_output= mydata.iloc[:,3]
```

```python
x_input.dtypes
```

```python
x_input.corr()
```

```python
#plotting histograms for each feature 
plt.figure(figsize=(16,16))
x_input.hist(bins=50)
plt.show()
```

```python
#plotting scatter plot for each feature v/s CO2 Emission to get the one having strongest correlation
plt.figure(figsize=(16,16))
i=1
for col in x_input:
    plt.subplot(3,3,i)
    plt.scatter(x_input[col],y_output)
    plt.title(col + 'v/s' + 'CO2_EMISSIOn')
    plt.grid()
    i=i+1
plt.show() 
```
